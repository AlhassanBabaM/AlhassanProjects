```python
# prompt: clean double quotation marks and duplicates from the data

import pandas as pd
def clean_data(data):
  cleaned_data = []
  for row in data:
    cleaned_row = []
    for item in row:
      cleaned_item = item.replace('"', '')
      cleaned_row.append(cleaned_item)
    cleaned_data.append(cleaned_row)

  # Remove duplicates
  cleaned_data = list(set(tuple(row) for row in cleaned_data))

  return cleaned_data

cleaned_data = clean_data(cleaned_data)

pd.DataFrame(cleaned_data)
```
